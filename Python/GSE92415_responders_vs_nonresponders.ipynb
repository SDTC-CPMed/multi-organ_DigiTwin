```python
import pandas as pd
import numpy as np
#import seaborn as sns
#import matplotlib.pyplot as plt
#from scipy.spatial.distance import pdist
#from seriate import seriate
```

```python
responders = pd.read_table('../data/GSE92415/UR_predictions_responders.txt', sep = ',', index_col = 0).transpose()
nonresponders = pd.read_table('../data/GSE92415/UR_predictions_nonresponders.txt', sep = ',', index_col = 0).transpose()
```

```python
specificity = list()
for sp in responders.columns:
    responders_sp = responders.index[np.where(responders[sp] < 0.05)[0]]
    nonresponders_sp = nonresponders.index[np.where(nonresponders[sp] < 0.05)[0]]
    specificity_sp = list()
    for gene in np.union1d(responders.index, nonresponders.index):
        if gene in np.intersect1d(responders_sp, nonresponders_sp):
            specificity_sp.append('Both')
        elif gene in responders_sp:
            specificity_sp.append('Responders')
        elif gene in nonresponders_sp:
            specificity_sp.append('Nonresponders')
        else:
            specificity_sp.append('Neither')
    specificity.append(specificity_sp)
```

```python
specificity = pd.DataFrame(specificity)
specificity.index = responders.columns
specificity.columns = np.union1d(responders.index, nonresponders.index)
specificity
```

```python
specificity.to_csv('../data/GSE92415/URs_RvsNR.csv', header = True, index = True)
```
